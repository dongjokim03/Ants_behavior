# Functions

```python
# Heaviside step function for defining localized influence zones
def H(r, pos, a):
    # Returns 1 if the distance from pos to r is within radius a, else 0
    return np.where(np.linalg.norm(r - pos, axis=1)**2 <= a**2, 1, 0)


# Apply periodic boundary conditions to agent positions
def apply_periodic_boundary(positions, L):
    # Maps coordinates to the range [-L, L] using modulus arithmetic
    positions[:, 0] = (positions[:, 0] + L) % (2 * L) - L
    positions[:, 1] = (positions[:, 1] + L) % (2 * L) - L
    return positions


# Compute the gradient of the concentration field c using finite differences
def compute_concentration_gradient(c, dx):
    dc_dx = np.gradient(c, axis=0) / dx  # Gradient in x-direction
    dc_dy = np.gradient(c, axis=1) / dx  # Gradient in y-direction
    return dc_dx, dc_dy


# Compute normal acceleration for each agent based on local concentration gradient
def compute_normal_acceleration(agent_positions, agent_directions, c, x, y, dx):
    # Compute spatial concentration gradients
    dc_dx, dc_dy = compute_concentration_gradient(c, dx)

    # Create interpolation functions for concentration gradients
    interp_dc_dx = RegularGridInterpolator((x, y), dc_dx)
    interp_dc_dy = RegularGridInterpolator((x, y), dc_dy)

    # Initialize acceleration array
    accel = np.zeros_like(agent_positions)

    # Loop over all agents
    for i, pos in enumerate(agent_positions):
        # Compute the agent's normal vector (perpendicular to heading direction)
        direction_vector = np.array([-np.sin(agent_directions[i]), np.cos(agent_directions[i])])

        # Interpolate concentration gradient at the agent's position
        grad_c = np.array([interp_dc_dx(pos), interp_dc_dy(pos)]).flatten()

        # Project gradient onto agent's normal direction
        grad_dot_dir = np.dot(grad_c, direction_vector)

        """
        Compute the normal acceleration using a mixture of functions:
        - Peaked aversion within specific range (threhold)
        - A thresholded hyperbolic tangent (sharp switch near threshold)
        """
        accel[i] = K * (a * -grad_dot_dir) * np.exp(-(a * grad_dot_dir)**2) + \
                   G_tanh * np.tanh(np.abs(grad_dot_dir) - th) * np.sign(grad_dot_dir) * np.heaviside(np.abs(grad_dot_dir) - th, 1)

    return accel


# Compute the time derivative of the concentration field (dc/dt)
def update_concentration(c, agent_positions, kp, km, N, dx):
    c = c.reshape((N, N))  # Reshape flat concentration field to 2D grid

    # Generate position grid (X, Y must be defined globally)
    r = np.stack((X, Y), axis=-1).reshape(-1, 2)

    # Initialize source field (sum of Heaviside kernels)
    H_sum = np.zeros_like(c)
    for pos in agent_positions:
        H_sum += H(r, pos, ph_r).reshape((N, N))  # Add contributions from each agent

    # Reaction-diffusion style update: production - decay
    dc_dt = kp * H_sum - km * c
    return dc_dt.flatten()


# Main function that defines the dynamics of agents and concentration over time
def agent_dynamics(t, s, n_agents, v0, N, dx, kp, km, noise_strength=1):
    # Parse the state vector:
    # positions: shape (n_agents, 2), directions: shape (n_agents,), c: shape (N, N)
    positions = s[:2 * n_agents].reshape(n_agents, 2)
    positions = apply_periodic_boundary(positions, L)  # Enforce periodic boundary conditions

    directions = s[2 * n_agents:3 * n_agents]  # Heading angles (theta)
    c = s[3 * n_agents:].reshape(N, N)         # Concentration field

    # Compute normal acceleration for each agent
    accel = compute_normal_acceleration(positions, directions, c, x, y, dx)

    # Derivative of heading angle (change in orientation)
    dtheta_dt = accel[:, 0]  # Take x-component (projected normal acceleration)
    # dtheta_dt = np.zeros_like(accel[:, 0])  # (optional: no reaction)

    # Add angular noise (Gaussian white noise)
    noise = np.random.normal(0, noise_strength, size=n_agents)
    dtheta_dt += noise

    # Compute position update based on fixed speed and direction
    dx_dt = v0 * np.cos(directions)
    dy_dt = v0 * np.sin(directions)
    dpos_dt = np.stack([dx_dt, dy_dt], axis=-1).reshape(-1)

    # Compute change in concentration field
    dc_dt_flat = update_concentration(c, positions, kp, km, N, dx)

    # Return time derivatives as a flattened state vector
    return np.concatenate([dpos_dt, dtheta_dt, dc_dt_flat])
```

```python
# Estimate the slope (fractal exponent) from log-log plot data
def calculate_slope(x_vals, y_vals, start_idx, end_idx):
    # Linear regression in log-log space to find scaling exponent
    coeffs = np.polyfit(np.log10(x_vals[start_idx:end_idx]), np.log10(y_vals[start_idx:end_idx]), 1)
    return coeffs[0]  # Return slope


# Compute Mean Squared Displacement (MSD) from the origin (t = 0)
def calculate_msd_from_origin(x, y):
    n_steps = len(x)
    msd = np.zeros(n_steps)
    x0, y0 = x[0], y[0]  # Initial position

    for t in range(n_steps):
        # Squared displacement from the origin at each time
        displacements = (x[t] - x0)**2 + (y[t] - y0)**2
        msd[t] = displacements

    return msd


# Reconstruct continuous trajectory from periodic positions
def reconstruct_continuous_trajectory(positions, L):
    # positions shape: [2n, T], even rows = x, odd rows = y
    continuous_positions = np.copy(positions)

    for i in range(1, positions.shape[1]):  # Loop over time
        for j in range(0, positions.shape[0], 2):  # Loop over agents (x/y paired)
            # Compute displacement between time steps
            delta_x = continuous_positions[j, i] - continuous_positions[j, i - 1]
            delta_y = continuous_positions[j + 1, i] - continuous_positions[j + 1, i - 1]

            # Apply periodic correction (unwrap displacement)
            delta_x = (delta_x + L) % (2 * L) - L
            delta_y = (delta_y + L) % (2 * L) - L

            # Update position by adding corrected displacement
            continuous_positions[j, i] = continuous_positions[j, i - 1] + delta_x
            continuous_positions[j + 1, i] = continuous_positions[j + 1, i - 1] + delta_y

    return continuous_positions  # Returns unwrapped trajectories

```

# Main

```python
# === Antennation (sensory-motor response) Parameters ===
K = 3.5           # Aversion intensity
a = 10.45       # Aversion threshold
G_tanh = 0.1      # Attraction intensity
th = 1.0          # Attraction threshold


# === Environmental Parameters ===
kp = 0.01       # Footprint concentration production rate
km = 0.2        # Footprint concentration decay rate
ph_r = 0.03     # Agent footprint radius
L = 0.5         # Half-size of 2D simulation domain (domain: [-L, L])
dx = ph_r / 3   # Grid spacing for concentration field
x = np.arange(-L, L + dx, dx)  # Discretized x-axis
y = np.arange(-L, L + dx, dx)  # Discretized y-axis
X, Y = np.meshgrid(x, y, indexing='ij')  # Create 2D meshgrid
N = len(x)      # Number of grid points per axis

# === Agent Parameters ===
n_agents = 65   # Number of agents
T_end = 200     # Simulation time (end)
v0 = 0.02       # Constant agent speed

# Time discretization
noise_strength = 0.5                      # Standard deviation of angular noise
t_span = (0, T_end)                       # Simulation time span
t_eval = np.arange(t_span[0], t_span[1]+1, 1)  # Evaluation points

# Record start time
tik = time.time()

# Initialize result storage arrays
density_gt_1_array = []       # Probability that density > 1
MSD_slopes = []               # MSD slope (diffusion exponent)

# Number of simulations to run
num_experiments = 100

# === Simulation loop ===
for i in [101]:  # You can change this to range(num_experiments) if needed
    np.random.seed(i)  # Set unique seed for each trial

    # Initialize agent positions (uniform in domain) and directions (uniform on circle)
    agent_positions = np.random.rand(n_agents, 2) * 2 * L - L
    agent_directions = np.random.rand(n_agents) * 2 * np.pi
    c0 = np.zeros((N, N))  # Initial concentration field (zeros)

    # Combine all into initial state vector
    s0 = np.concatenate([agent_positions.flatten(), agent_directions, c0.flatten()])

    # Solve dynamics using ODE solver
    sol = solve_ivp(agent_dynamics, t_span, s0,
                    args=(n_agents, v0, N, dx, kp, km, noise_strength),
                    max_step=0.1, t_eval=t_eval)
    
    s = sol.y  # Extract solution (shape: [state_dim, time])
    positions_over_time = s[:2 * n_agents].reshape(2 * n_agents, -1)
    directions_over_time = s[2 * n_agents:3 * n_agents]

    # Apply periodic boundary correction to recover continuous trajectories
    contd_positions = reconstruct_continuous_trajectory(positions_over_time, L)

    ### --- Compute Density (ρ > 1) --- ###
    density_list = []
    for t_idx in range(len(t_eval)):
        x_t = positions_over_time[0::2, t_idx]
        y_t = positions_over_time[1::2, t_idx]
        # Count how many neighbors are within ph_r for each agent
        density = np.array([
            np.sum(np.sqrt((x_t - x_t[idx])**2 + (y_t - y_t[idx])**2) <= ph_r)
            for idx in range(n_agents)
        ])
        density_list.append(density)
    
    density_array = np.array(density_list)
    selected_density = density_array.flatten()

    # Compute histogram for density values
    bin_edges = np.arange(-0.5, np.max(selected_density) + 1.5, 1)
    counts, _ = np.histogram(selected_density, bins=bin_edges, density=True)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

    # Probability that local density exceeds 1
    probability_gt_1 = np.sum(counts[bin_centers > 1])

    ### --- MSD (Mean Squared Displacement) --- ###
    all_msds = []
    for sel_number in range(n_agents):
        contd_positions_sel_x = contd_positions[sel_number * 2, :]
        contd_positions_sel_y = contd_positions[sel_number * 2 + 1, :]
        msd = calculate_msd_from_origin(contd_positions_sel_x, contd_positions_sel_y)
        all_msds.append(msd)

    mean_msd = np.mean(all_msds, axis=0)

    # Compute slope (diffusion exponent) from selected time window
    start_idx = 15
    end_idx = 100
    msd_slope = calculate_slope(t_eval[start_idx:end_idx], mean_msd[start_idx:end_idx], start_idx, end_idx)

    # Save results
    density_gt_1_array.append(probability_gt_1)
    MSD_slopes.append(msd_slope)

    print(f"Experiment random_seed={i} completed")

# === Print Final Results ===
print(f"Density > 1: {density_gt_1_array}")
print(f"MSD Slopes: {MSD_slopes}")
print('Time: {:.4f}s'.format(time.time() - tik))
```

# Visualization

```python
# === Prepare data for visualization ===
s = sol.y  # Extract full simulation state from ODE solver output
sel_number = 0  # Index of selected agent to highlight

# Reshape simulation state: positions, directions, concentration field
positions_over_time = s[:2 * n_agents].reshape(2 * n_agents, -1)
directions_over_time = s[2 * n_agents:3 * n_agents]
concentration_over_time = s[3 * n_agents:].reshape(N, N, -1)

# Extract position and direction of selected agent over time
positions_sel_x = positions_over_time[2 * sel_number, :]
positions_sel_y = positions_over_time[2 * sel_number + 1, :]
directions_sel = s[2 * n_agents + sel_number, :]

# === Define custom colormap for concentration visualization ===
colors = [
    np.array([255, 250, 245]) / 255,   # White
    np.array([255, 179, 141]) / 255,   # Orange
    np.array([234, 85, 20]) / 255      # Red
]
positions = [0, 0.5, 1]  # Corresponding color stops in gradient

# Create custom color map using white-orange-red transition
custom_cmap = LinearSegmentedColormap.from_list("WhiteOrangeRed", list(zip(positions, colors)))

# Define color normalization and mapping for footprint concentration
vmax = np.max(concentration_over_time)
vmin = 0
norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)
colormapping = cm.ScalarMappable(norm=norm, cmap=custom_cmap)

# === Initialize matplotlib figure ===
plt.rc('font', size=16, family='Arial')
fig, ax = plt.subplots(1, 1, figsize=(5.5, 5))  # Single subplot

# === Animation update function ===
def animate(i):
    window = L  # Size of the visible area (from -L to L)
    ax.clear()  # Clear previous frame

    # If time index exceeds simulation range, hold last frame
    if i >= len(t_eval):
        i = len(t_eval) - 1

    # Plot footprint concentration field at current time
    c = concentration_over_time[:, :, i]
    ax.contourf(X, Y, c, 100, cmap=custom_cmap, norm=norm)

    # Get agent positions and directions at current time
    positions = positions_over_time[:, i].reshape(n_agents, 2)
    dx = v0 * np.cos(directions_over_time[:, i])  # Velocity x-component
    dy = v0 * np.sin(directions_over_time[:, i])  # Velocity y-component

    # Plot agents as black dots
    ax.scatter(positions[:, 0], positions[:, 1], s=50, alpha=0.4, color='k')

    # Plot direction vectors (velocity arrows)
    ax.quiver(positions[:, 0], positions[:, 1], dx, dy,
              color='b', alpha=0.4, scale_units='xy', angles='xy',
              scale=0.25, width=0.005, headwidth=5)

    # Plot selected agent: current position and past trajectory
    ax.scatter(positions_sel_x[i], positions_sel_y[i], s=50, color='k')  # current position
    ax.scatter(positions_sel_x[:i], positions_sel_y[:i], marker='o', s=10, color='r')  # trajectory history
    ax.quiver(positions_sel_x[i], positions_sel_y[i], dx[sel_number], dy[sel_number],
              color='b', alpha=1, scale_units='xy', angles='xy',
              scale=0.25, width=0.005, headwidth=5)  # current velocity vector

    # Set title and axis properties
    ax.set_title(f'Time: {t_eval[i]:.2f}')
    ax.set_xticks(np.arange(-1, 1, 0.5))
    ax.set_yticks(np.arange(-1, 1, 0.5))
    ax.set_xlim(-window, window)
    ax.set_ylim(-window, window)
    ax.set_aspect('equal')  # Keep aspect ratio square

# === Colorbar and layout ===
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
cbar = fig.colorbar(colormapping, ax=ax, pad=0.02, fraction=0.046, shrink=1)
cbar.set_ticks([0, 0.1])

# Set transparent background for figure and axes
fig.patch.set_alpha(0)
ax.patch.set_alpha(0)

# === Create animation ===
anim = animation.FuncAnimation(fig, animate, frames=len(t_eval) + 3)

# Save animation (commented out by default)
anim.save('Results/Simulation_trajectory_K'+str(K)+'_a'+str(a)+'_G'+str(G_tanh)+'_th'+str(th)+'.mp4', 
          fps=10, dpi=240)

# Render animation as HTML5 video for Jupyter notebooks
HTML(anim.to_html5_video())
```

```python
# Define the radius for local density calculation
density_grain = ph_r  # Agents within this radius are considered neighbors

# Initialize a list to store density values over time
density_list = []

# Loop through each time frame in the simulation
for i in range(len(t_eval)):
    # Extract agent positions at current time (x and y coordinates)
    x_t = positions_over_time[0::2, i]  # x positions of all agents
    y_t = positions_over_time[1::2, i]  # y positions of all agents

    # Initialize density array for this time frame
    density = np.zeros(n_agents)
    for agent_idx in range(n_agents):
        # Compute pairwise distances to all other agents
        distances = np.sqrt((x_t[agent_idx] - x_t)**2 + (y_t[agent_idx] - y_t)**2)
        # Count how many agents (including self) are within the density radius
        density[agent_idx] = np.sum(distances <= density_grain)

    # Store computed density values
    density_list.append(density)

# Convert list to NumPy array (shape: [n_time_frames, n_agents])
density_array = np.array(density_list)

# Flatten the array to consider all time and agents for PDF calculation
selected_density = density_array.flatten()

# === Compute Probability Density Function (PDF) ===
min_density = np.min(selected_density)
max_density = np.max(selected_density)

# Create bin edges and centers for histogram
bin_edges = np.arange(-0.5, max_density + 1.5, 1)  # Integer binning
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

# Calculate PDF (normalized histogram)
counts, _ = np.histogram(selected_density, bins=bin_edges, density=True)

# Calculate probability that density > 1
mask_gt_1 = bin_centers > 1
probability_gt_1 = np.sum(counts[mask_gt_1])
probability_text = f"Prob.($\\rho > 1$) = {probability_gt_1:.4f}"

# === Plot PDF as a line chart ===
plt.figure(figsize=(6, 4))
plt.plot(bin_centers, counts, 'o-', linewidth=2, markersize=10, color='tab:green', label=probability_text)

plt.xlabel('Density')
plt.ylabel('Probability')
plt.xticks(np.arange(0, 16, 1))
plt.yticks(np.arange(0, 1.2, 0.2))
plt.xlim(0, 6)
plt.ylim(0, 1)

# Add legend with computed probability
plt.legend(loc='upper right', fontsize=16)

# Save figure as a high-resolution image
plt.savefig('Results/Simulation_Aggregation_Probability_K'+str(K)+'_a'+str(a)+'_G'+str(G_tanh)+'_th'+str(th)+'.jpg', 
            bbox_inches='tight', dpi=300)

plt.show()

# Print numeric result of density > 1 probability
print(f"Probability that density > 1: {probability_gt_1:.4f}")
```

```python
# === Compute Mean Squared Displacement (MSD) for all agents ===
msd_all_agents = []

# Loop through all agents to compute MSD individually
for sel_number in range(n_agents):
    temp_x = contd_positions[2 * sel_number, :]        # x trajectory
    temp_y = contd_positions[2 * sel_number + 1, :]    # y trajectory

    # Compute MSD from the origin for each agent
    msd = calculate_msd_from_origin(temp_x, temp_y)
    msd_all_agents.append(msd)

# Convert list to array for statistical analysis
msd_all_agents = np.array(msd_all_agents)

# Compute average and standard deviation of MSD across agents
mean_msd = np.mean(msd_all_agents, axis=0)
std_msd = np.std(msd_all_agents, axis=0)

# Use the simulation time points as x-axis
time_steps = t_eval

# === Compute scaling exponent β from log-log slope of MSD curve ===
start_idx = 15     # Start of log-log fitting window
end_idx = 100      # End of log-log fitting window
slope = calculate_slope(time_steps, mean_msd, start_idx, end_idx)

# Output the computed slope (diffusion exponent)
print(f"Calculated slope (β) in 15≤T≤100: {slope:.2f}")

# === Plot MSD curve with reference slopes ===
plt.figure(figsize=(6, 4))

# Optional: shaded region for standard deviation
# plt.fill_between(time_steps[1:], mean_msd[1:] - std_msd[1:], mean_msd[1:] + std_msd[1:], color='gray', alpha=0.3, label='±Std. Dev.')

# Plot mean MSD
# plt.scatter(time_steps[1:], mean_msd[1:], s=25, color='k', label=f'Mean MSD, β = {slope:.2f} @15<t<100', zorder=1)
plt.plot(time_steps[1:], mean_msd[1:], lw=3, color='k', label='Mean MSD', zorder=1)

# Plot reference line with slope = 2 (ballistic motion)
msd_line_2 = mean_msd[1] * (time_steps[1:] / time_steps[1])**2
plt.plot(time_steps[1:], msd_line_2, 'b--', label='Slope = 2', zorder=2)

# Plot reference line with slope = 1 (normal diffusion)
msd_line_1 = mean_msd[1] * (time_steps[1:] / time_steps[1])**1
plt.plot(time_steps[1:], msd_line_1, 'r--', label='Slope = 1', zorder=2)


# Set log-log scale and plot limits
plt.xscale('log')
plt.yscale('log')
plt.xlim(0.9, 200)
plt.ylim(2e-4, 20)

# Axis labels and legend
plt.xlabel('Time (t)')
plt.ylabel('MSD')
plt.legend(loc='best', fontsize=14)
# plt.grid(True)

# Save plot as high-resolution image
plt.savefig('Results/Simulation_MSD_K' + str(K) + '_a' + str(a) + '_G' + str(G_tanh) + '_th' + str(th) + '.jpg',
            bbox_inches='tight', dpi=300)

plt.show()
```
